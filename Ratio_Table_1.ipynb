```python
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
columns = [
    'id',
    'name',
    'short_name',
    'year',
    'indicator',
    'numerator',
    'denominator'
]

age_group_mapping = {
    'всего': '1',
    '0-4': '4',
    '5-6': '5',
    '7-14': '6',
    '15-17': '7',
    '18-24': '8',
    '25-34': '9',
    '35-44': '10',
    '45-54': '11',
    '55-64': '12',
    '65+': '13'
}

data = []
data_2016_2018 = []

df_table_1 = pd.DataFrame(columns=columns)
df_table_1_2016_2018 = pd.DataFrame(columns=columns)

for row_number in range(0, df_population.shape[0]): # 2016-2024
# for row_number in range(0, 9): # только 2016-2018
     for col_number in range(2,13):
        new_row = []

        age_group = df_population.columns[col_number]
        base_id = age_group_mapping[age_group]

        if df_population.iloc[row_number, 1] == 'всего':
            id_value = f'id_1.{base_id}'
        elif df_population.iloc[row_number, 1] == 'мужского населения':
            id_value = f'id_1.{base_id}.1'
        else:  # женское население
            id_value = f'id_1.{base_id}.2'
        # display(id_value)

        if id_value == 'id_1.1.1':
            id_value = 'id_1.2'
        if id_value == 'id_1.1.2':
            id_value = 'id_1.3'
            
        new_row.append(id_value)
         
        # new_row.append('id_1.1') # переделать
         
        if df_population.iloc[row_number, 1] == 'всего':
            if df_population.columns[col_number] == 'всего':
                new_row.append(f'Показатель первичной заболеваемости туберкулезом {df_population.columns[col_number]} (на 100 тыс. населения)')
                new_row.append(f'ПЗ туб тер')
            elif (df_population.columns[col_number] == '5-6') or (df_population.columns[col_number] == '7-14') or (df_population.columns[col_number] == '15-17') or (df_population.columns[col_number] == '65+'):
                new_row.append(f'Показатель первичной заболеваемости туберкулезом {df_population.columns[col_number]} лет (на 100 тыс. населения)')
                new_row.append(f'ПЗ туб {df_population.columns[col_number]}')
            else:
                new_row.append(f'Показатель первичной заболеваемости туберкулезом {df_population.columns[col_number]} года (на 100 тыс. населения)')
                new_row.append(f'ПЗ туб {df_population.columns[col_number]}')
        else:
            if df_population.columns[col_number] == 'всего':
                new_row.append(f'Показатель первичной заболеваемости туберкулезом {df_population.iloc[row_number, 1]} {df_population.columns[col_number]} (на 100 тыс. населения)')
                new_row.append(f'ПЗ туб {df_population.iloc[row_number, 1][:3]}')
            elif (df_population.columns[col_number] == '5-6') or (df_population.columns[col_number] == '7-14') or (df_population.columns[col_number] == '15-17') or (df_population.columns[col_number] == '65+'):
                new_row.append(f'Показатель первичной заболеваемости туберкулезом {df_population.iloc[row_number, 1]} {df_population.columns[col_number]} лет (на 100 тыс. населения)')
                new_row.append(f'ПЗ туб {df_population.iloc[row_number, 1][:3]} {df_population.columns[col_number]}')
            else:
                new_row.append(f'Показатель первичной заболеваемости туберкулезом {df_population.iloc[row_number, 1]} {df_population.columns[col_number]} года (на 100 тыс. населения)')
                new_row.append(f'ПЗ туб {df_population.iloc[row_number, 1][:3]} {df_population.columns[col_number]}')
           
        new_row.append(df_population.iloc[row_number, 0]) # год


        if df_population.iloc[row_number, 1] == 'всего': 
        #форма 8, табл.1000,   стр.1+2,   гр.5) / Среднегодовая численность населения Иркутской области* 100000
            value1 = df_1000[(df_1000['Год'] == df_population.iloc[row_number, 0]) & (df_1000['Строка'] == 1) & (df_1000['Графа'] == col_number + 3)]['Значение'].iloc[0]
            value2 = df_1000[(df_1000['Год'] == df_population.iloc[row_number, 0]) & (df_1000['Строка'] == 2) & (df_1000['Графа'] == col_number + 3)]['Значение'].iloc[0]
            numerator = value1 + value2
        elif df_population.iloc[row_number, 1] == 'мужского населения':
            numerator = df_1000[(df_1000['Год'] == df_population.iloc[row_number, 0]) & (df_1000['Строка'] == 1) & (df_1000['Графа'] == col_number + 3)]['Значение'].iloc[0]
        else:
            numerator = df_1000[(df_1000['Год'] == df_population.iloc[row_number, 0]) & (df_1000['Строка'] == 2) & (df_1000['Графа'] == col_number + 3)]['Значение'].iloc[0]

        denominator = df_population.iloc[row_number, col_number]
    
        indicator = numerator * 100000/denominator
    
        new_row.append(round(indicator, 2))
        new_row.append(numerator)
        new_row.append(denominator)
        data.append(new_row)
        if df_population.iloc[row_number, 0] <= 2018:
            data_2016_2018.append(new_row)

df_table_1 = pd.DataFrame(data, columns=columns)
df_table_1_2016_2018 = pd.DataFrame(data_2016_2018, columns=columns)
```
