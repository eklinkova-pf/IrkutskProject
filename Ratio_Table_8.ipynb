```python
path = 'E:/IrkutskProject/Data/'

df_1100 = pd.read_csv(f'{path}Common_2016-2018/1100_2016-2018.csv')
# df_population = pd.read_csv(f'{path}Common_2016-2018/avg_population_2016-2018.csv')
df_population = pd.read_csv(f'{path}Common/avg_population_2016-2024.csv')
```

```python
years = [2016, 2017, 2018]

columns = [
    'id',
    'name',
    'short_name',
    'year',
    'indicator',
    'numerator',
    'denominator'
]

data = []

for year in years:
    new_row = []
    new_row.append('id_8.1')
    new_row.append('Обеспеченность населения фтизиатрами (на 10 000 населения)')
    new_row.append('Врачи туб')
    new_row.append(year)
    
    # форма 30, табл.1100,   стр.111,   гр.3) /    Среднегодовая численность населения Иркутской области  * 10000
    numerator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 110) & (df_1100['Графа'] == 3)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['Всего'].iloc[0]

    if denominator != 0:
        indicator = round(numerator*10000/denominator, 2)
    else:
        indicator = 0.00

    new_row.append(indicator)
    new_row.append(numerator)
    new_row.append(denominator)

    data.append(new_row)
    #----
    new_row = []
    new_row.append('id_8.2')
    new_row.append('Укомплектованность кадрами (%)')
    new_row.append('УК врачи туб')
    new_row.append(year)
    
    # форма 30, табл.1100,   стр.111,   гр.4) /     форма 30, табл.1100,   стр.111,   гр.3) * 100%
    numerator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 110) & (df_1100['Графа'] == 4)]['Значение'].iloc[0]
    denominator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 110) & (df_1100['Графа'] == 3)]['Значение'].iloc[0]

    if denominator != 0:
        indicator = round(numerator*100/denominator, 2)
    else:
        indicator = 0.00

    new_row.append(indicator)
    new_row.append(numerator)
    new_row.append(denominator)

    data.append(new_row)
    #----
    new_row = []
    new_row.append('id_8.3')
    new_row.append('Укомплектованность физическими лицами (%)')
    new_row.append('УФ врачи туб')
    new_row.append(year)
    
    # форма 30, табл.1100,   стр.111,   гр.9) /     форма 30, табл.1100,   стр.111,   гр.3) * 100%
    numerator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 110) & (df_1100['Графа'] == 9)]['Значение'].iloc[0]
    denominator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 110) & (df_1100['Графа'] == 3)]['Значение'].iloc[0]

    if denominator != 0:
        indicator = round(numerator*100/denominator, 2)
    else:
        indicator = 0.00

    new_row.append(indicator)
    new_row.append(numerator)
    new_row.append(denominator)

    data.append(new_row)
    #----
    new_row = []
    new_row.append('id_8.4')
    new_row.append('Коэффициент совместительства')
    new_row.append('КС врачи туб')
    new_row.append(year)
    
    # форма 30, табл.1100,   стр.111,   гр.4) /     форма 30, табл.1100,   стр.111,   гр.9)
    numerator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 110) & (df_1100['Графа'] == 4)]['Значение'].iloc[0]
    denominator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 110) & (df_1100['Графа'] == 9)]['Значение'].iloc[0]

    if denominator != 0:
        indicator = round(numerator/denominator, 2)
    else:
        indicator = 0.00

    new_row.append(indicator)
    new_row.append(numerator)
    new_row.append(denominator)

    data.append(new_row)
    #----
    new_row = []
    new_row.append('id_8.5')
    new_row.append('Обеспеченность населения фтизиатрами участковыми (на 10 000 населения)')
    new_row.append('Врачи туб участ')
    new_row.append(year)
    
    # форма 30, табл.1100,   стр.112,   гр.3) /    Среднегодовая численность населения Иркутской области  * 10000
    numerator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 111) & (df_1100['Графа'] == 3)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['Всего'].iloc[0]

    if denominator != 0:
        indicator = round(numerator*10000/denominator, 2)
    else:
        indicator = 0.00

    new_row.append(indicator)
    new_row.append(numerator)
    new_row.append(denominator)

    data.append(new_row)
    #----
    new_row = []
    new_row.append('id_8.6')
    new_row.append('Укомплектованность кадрами  фтизиатрами участковыми (%)')
    new_row.append('УК врачи туб участ')
    new_row.append(year)
    
    # форма 30, табл.1100,   стр.112,   гр.4) /     форма 30, табл.1100,   стр.112,   гр.3) * 100%
    numerator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 111) & (df_1100['Графа'] == 4)]['Значение'].iloc[0]
    denominator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 111) & (df_1100['Графа'] == 3)]['Значение'].iloc[0]

    if denominator != 0:
        indicator = round(numerator*100/denominator, 2)
    else:
        indicator = 0.00

    new_row.append(indicator)
    new_row.append(numerator)
    new_row.append(denominator)

    data.append(new_row)

    #----
    new_row = []
    new_row.append('id_8.7')
    new_row.append('Укомплектованность  фтизиатрами участковыми физическими лицами (%)')
    new_row.append('УФ врачи туб участ')
    new_row.append(year)
    
    # форма 30, табл.1100,   стр.112,   гр.9) /     форма 30, табл.1100,   стр.112,   гр.3) * 100%
    numerator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 111) & (df_1100['Графа'] == 9)]['Значение'].iloc[0]
    denominator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 111) & (df_1100['Графа'] == 3)]['Значение'].iloc[0]

    if denominator != 0:
        indicator = round(numerator*100/denominator, 2)
    else:
        indicator = 0.00

    new_row.append(indicator)
    new_row.append(numerator)
    new_row.append(denominator)

    data.append(new_row)
    
    #----
    new_row = []
    new_row.append('id_8.8')
    new_row.append('Коэффициент совместительства  фтизиатрами участковыми')
    new_row.append('КС врачи туб участ')
    new_row.append(year)
    
    # форма 30, табл.1100,   стр.112,   гр.4) /     форма 30, табл.1100,   стр.112,   гр.9)
    numerator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 111) & (df_1100['Графа'] == 4)]['Значение'].iloc[0]
    denominator = df_1100[(df_1100['Год'] == year) & (df_1100['Строка'] == 111) & (df_1100['Графа'] == 9)]['Значение'].iloc[0]

    if denominator != 0:
        indicator = round(numerator/denominator, 2)
    else:
        indicator = 0.00

    new_row.append(indicator)
    new_row.append(numerator)
    new_row.append(denominator)

    data.append(new_row)
    
    
df_table_8 = pd.DataFrame(data, columns=columns)    
```
