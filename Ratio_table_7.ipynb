```python
# Загрузка библиотек
import pandas as pd
import numpy as np
```

```python
path = 'E:/IrkutskProject/Data/'

df_population = pd.read_csv(f'{path}Common/avg_population_2016-2024.csv')
df_2513 = pd.read_csv(f'{path}Common/2513_2016-2024.csv')
```

```python
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
columns = [
    'id',
    'name',
    'short_name',
    'year',
    'indicator',
    'numerator',
    'denominator'
]

data = []
data_2016_2018 = []

df_table_7 = pd.DataFrame(columns=columns)
df_table_7_2016_2018 = pd.DataFrame(columns=columns)

for year in years:
    new_row = []
    # id_7.1 форма 30, табл.2513,   стр.1,   гр.3) /    Среднегодовая численность населения Иркутской области  * 100%
    new_row.append('id_7.1')
    new_row.append('Охват населения всеми видами осмотров (в %)')
    new_row.append('Охват проф.осмотр')
    new_row.append(year)
    numerator = df_2513[(df_2513['Год'] == year) & (df_2513['Строка'] == 1) & (df_2513['Графа'] == 3)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['Всего'].iloc[0]
    indicator = numerator*100/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)

#-----------------
    new_row = []
    # id_7.2 форма 30, табл.2513,   стр.4+5,   гр.3) /    Среднегодовая численность населения Иркутской области  в возрасте 0-14 лет * 100%
    new_row.append('id_7.2')
    new_row.append('Охват иммунодиагностикой детей 0-14 лет (в %)')
    new_row.append('Охват иммунол. 0-14')
    new_row.append(year)
    value_1 = df_2513[(df_2513['Год'] == year) & (df_2513['Строка'] == 4) & (df_2513['Графа'] == 3)]['Значение'].iloc[0]
    value_2 = df_2513[(df_2513['Год'] == year) & (df_2513['Строка'] == 5) & (df_2513['Графа'] == 3)]['Значение'].iloc[0]
    numerator = (
        (value_1 if pd.notnull(value_1) else 0) +
        (value_2 if pd.notnull(value_2) else 0)
    )
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['0-14'].iloc[0]
    indicator = numerator*100/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
#-----------------
    new_row = []
    # id_7.3 форма 30, табл.2513,   стр.2,   гр.3) /    Среднегодовая численность населения Иркутской области  старше 15 лет * 100%
    new_row.append('id_7.3')
    new_row.append('Охват флюорографическими осмотрами населения старше 15 лет (в %)')
    new_row.append('Охват ФЛГ')
    new_row.append(year)
    numerator = (
        df_2513[(df_2513['Год'] == year) & (df_2513['Строка'] == 2) & (df_2513['Графа'] == 3)]['Значение'].iloc[0]
    )
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['15+'].iloc[0]
    indicator = numerator*100/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
#-----------------
    new_row = []
    # id_7.4 форма 30, табл.2513,   стр.2,   гр.5) /     форма 30, табл.2513,   стр.2,   гр.3)  * 1000
    new_row.append('id_7.4')
    new_row.append('Выявляемость туберкулеза методом флюорографии (1000 осмотренных)')
    new_row.append('Выявляемость ФЛГ')
    new_row.append(year)
    value_1 = df_2513[(df_2513['Год'] == year) & (df_2513['Строка'] == 2) & (df_2513['Графа'] == 5)]['Значение'].iloc[0]
    value_2 = df_2513[(df_2513['Год'] == year) & (df_2513['Строка'] == 2) & (df_2513['Графа'] == 3)]['Значение'].iloc[0]
    numerator = (value_1 if pd.notnull(value_1) else 0)
    denominator = (value_2 if pd.notnull(value_1) else 0)
    if denominator != 0:
        indicator = numerator*1000/denominator
    else:
        indicator = 0.0
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
#------------------------
    new_row = []
    # id_7.5 форма 30, табл.2513,   стр.3,   гр.5) /     форма 30, табл.2513,   стр.3,   гр.3)  * 1000
    new_row.append('id_7.5')
    new_row.append('Выявляемость туберкулеза бактериоскопически (1000 осмотренных)')
    new_row.append('Выявляемость бак')
    new_row.append(year)
    value_1 = df_2513[(df_2513['Год'] == year) & (df_2513['Строка'] == 3) & (df_2513['Графа'] == 5)]['Значение'].iloc[0]
    value_2 = df_2513[(df_2513['Год'] == year) & (df_2513['Строка'] == 3) & (df_2513['Графа'] == 3)]['Значение'].iloc[0]
    numerator = (value_1 if pd.notnull(value_1) else 0)
    denominator = (value_2 if pd.notnull(value_1) else 0)
    if denominator != 0:
        indicator = numerator*1000/denominator
    else:
        indicator = 0.0
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)

df_table_7 = pd.DataFrame(data, columns=columns)
df_table_7_2016_2018 = pd.DataFrame(data_2016_2018, columns=columns)
df_table_7
    
```
