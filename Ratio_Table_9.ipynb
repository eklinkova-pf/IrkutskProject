```python
path = 'E:/IrkutskProject/Data/'

df_population = pd.read_csv(f'{path}Common/avg_population_2016-2024.csv')
df_3100 = pd.read_csv(f'{path}Common/3100_2016-2024.csv')
```

```python
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
columns = [
    'id',
    'name',
    'short_name',
    'year',
    'indicator',
    'numerator',
    'denominator'
]

data = []
data_2016_2018 = []

df_table_9 = pd.DataFrame(columns=columns)
df_table_9_2016_2018 = pd.DataFrame(columns=columns)

for year in years:
    new_row = []
    # id_9.1 форма 30, табл.3100,   стр.57,   гр.5) /    Среднегодовая численность населения Иркутской области  * 10000
    new_row.append('id_9.1')
    new_row.append('Обеспеченность населения туберкулезными койками  для взрослых (на 10 000 населения)')
    new_row.append('Койки туб вз')
    new_row.append(year)
    numerator = df_3100[(df_3100['Год'] == year) & (df_3100['Строка'] == 57) & (df_3100['Графа'] == 5)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['18+'].iloc[0]
    #print('year = ', year, 'numerator = ', numerator, 'denominator =', denominator)
    indicator = numerator* 10000/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)

#---------------------
    new_row = []
    # id_9.2 форма 30, табл.3100,   стр.58,   гр.5) /    Среднегодовая численность детского населения (0-17 лет) Иркутской области  * 10000
    new_row.append('id_9.2')
    new_row.append('Обеспеченность населения туберкулезными койками  для детей (на 10 000 населения)')
    new_row.append('Койки туб дет')
    new_row.append(year)
    numerator = df_3100[(df_3100['Год'] == year) & (df_3100['Строка'] == 58) & (df_3100['Графа'] == 5)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['0-17'].iloc[0]
    #print('year = ', year, 'numerator = ', numerator, 'denominator =', denominator)
    indicator = numerator* 10000/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
        
#-------------------------
    new_row = []
    # id_9.3 форма 30, табл.3100,   стр.57,   гр.6 /    Среднегодовая численность взрослого (18 и старше)населения Иркутской области  * 100000
    new_row.append('id_9.3')
    new_row.append('Госпитализированная заболеваемость взрослого населения (на 100 тыс. населения)')
    new_row.append('ГЗ взр')
    new_row.append(year)
    numerator = df_3100[(df_3100['Год'] == year) & (df_3100['Строка'] == 57) & (df_3100['Графа'] == 6)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['18+'].iloc[0]
    #print('year = ', year, 'numerator = ', numerator, 'denominator =', denominator)
    indicator = numerator* 100000/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
        
#------------------------
    new_row = []
    # id_9.4 форма 30, табл.3100,   стр.58,   гр.6) /    Среднегодовая численность детского населения (0-17 лет) Иркутской области  * 100000
    new_row.append('id_9.4')
    new_row.append('Госпитализированная заболеваемость детского населения 0-17 (на 100 тыс. населения)')
    new_row.append('ГЗ дети 0-17')
    new_row.append(year)
    numerator = df_3100[(df_3100['Год'] == year) & (df_3100['Строка'] == 58) & (df_3100['Графа'] == 6)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['0-17'].iloc[0]
    #print('year = ', year, 'numerator = ', numerator, 'denominator =', denominator)
    indicator = numerator* 100000/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)

  
df_table_9 = pd.DataFrame(data, columns=columns)
df_table_9_2016_2018 = pd.DataFrame(data_2016_2018, columns=columns)
df_table_9
```
