1) Подкотовка датасетов формы 2300 за 2016-2018 года

```python
# Список годов для обработки
years = [2016, 2017, 2018]

# Название оригинального файла
original_names = {
    2016: 'ф33  правл за 2016г Иркутская обл.xlsx',
    2017: 'ф.33 и ф.8 Иркутская область 2017г..xls',
    2018: 'ф.33 и ф.8  Иркутская область  2018.xls'
}
original_url = {}
path = 'E:/IrkutskProject/Data/'

df_2400 = {}

for year in years:
    original_path = f'{path}{year}/Original/'
    original_url[year] = original_path + original_names[year]
    
    if year == 2016:
        df = pd.read_excel(original_url[year], engine='openpyxl', sheet_name = '2320,2330,2400')

        # Отфильтруем таблицу 2400
        df_2400[year] = df.iloc[9:27]
    else:
        df = pd.read_excel(original_url[year], engine='xlrd', sheet_name='2400-2500')
        df_2400[year] = df.iloc[0:16]
        df_2400[year] = df_2400[year].drop(df_2400[year].columns[1:8], axis=1)
        
    df_2400[year] = df_2400[year].dropna(how='all', axis=1)
    df_2400[year] = df_2400[year].drop(df_2400[year].columns[1:2], axis=1)
    df_2400[year] = df_2400[year].iloc[:, :8]
    df_2400[year] = df_2400[year].dropna(how='all')
    df_2400[year].drop(df_2400[year].index[1:2], inplace=True)
    
    df_2400[year].columns = [0, 1, 2, 3, 4, 5, 6, 7]
    df_2400[year] = df_2400[year].reset_index(drop=True)

    # Метки в соотвтетсвии с шаблоном
    df_2400[year].iloc[1, 0] = 'Взрослые, из них нуждающиеся в определении активности туберкулезного процесса (0А)'
    df_2400[year].iloc[2, 0] = 'Взрослые, из них нуждающиеся в проведении дифференциально-диагностических мероприятий (0Б)'
    df_2400[year].iloc[3, 0] = 'Взрослые с неактивным туб процессом после канонического излечения (III)'
    df_2400[year].iloc[4, 0] = 'Взрослые, состоящие в бытовом и производственном контакте с бактериовыделителем (IVA)'
    df_2400[year].iloc[5, 0] = 'Взрослые, состоящие в бытовом и производственном контакте с больным туберкулезом без бактериовыделения (IVA)'
    df_2400[year].iloc[6, 0] = 'Взрослые, состоящие в профессиональном контакте с источником инфекции (IVБ)'
    df_2400[year].iloc[7, 0] = 'Дети от 0 до 17 лет, из них нуждающиеся в уточнении характера туберкулезной чувствительности, уточнении активности туберкулеза и диагностике (0)'
    df_2400[year].iloc[8, 0] = 'Дети от 0 до 17 лет с остаточными посттуберкулезными изменениями (IIIА)'
    df_2400[year].iloc[9, 0] = 'Дети от 0 до 17 лет переведнные из I, II, IIIA групп (IIIБ)'
    df_2400[year].iloc[10, 0] = 'Дети от 0 до 17 лет, состоящие  в контакте с бактериовыделителями (IVA)'
    df_2400[year].iloc[11, 0] = 'Дети от 0 до 17 лет из контакта с больными туберкулезом без бактериовыделения, из семей животноводов или имеющих больных туберкулезом животных (IVБ)'
    df_2400[year].iloc[12, 0] = 'Дети от 0 до 17 лет в раннем периоде первичной туберкулезной инфекции (VIA)'
    df_2400[year].iloc[13, 0] = 'Дети от 0 до 17 лет ранее инфецированные, с гиперергической реакцией на туберкулин из соц групп риска с выраженными реакциями на туберкулин (VIБ)'
    df_2400[year].iloc[14, 0] = 'Дети от 0 до 17 лет с усиливающейся туберкулиновой чувствительностью (VIB)'

    df_2400[year].iloc[0, 1] = 'Взято на учет в текущем году'
    df_2400[year].iloc[0, 2] = 'Подлежало ХП или пробному лечению'
    df_2400[year].iloc[0, 3] = 'Прошел курс ХП, пробного лечения'
    df_2400[year].iloc[0, 4] = 'Впервые выявлено больных с ТБ'
    df_2400[year].iloc[0, 5] = 'Снято с учета'
    df_2400[year].iloc[0, 6] = 'Выбыло'
    df_2400[year].iloc[0, 7] = 'Состоит на конец года'

    df_2400[year].to_csv(f'E:/IrkutskProject/Data/{year}/2400.csv')
```

```python
columns = [
    'Показатель',
    'Уточнение',
    'Год',
    'Значение'
]

data = []

df_2400_long = pd.DataFrame(columns=columns)

for year in years:
    for row_number in range(1, df_2400[year].shape[0]):
        for col_number in range(1, 8):
            new_row = []
            new_row.append(df_2400[year].iloc[row_number, 0]) 
            new_row.append(df_2400[year].iloc[0, col_number])
            new_row.append(year)
            new_row.append(df_2400[year].iloc[row_number, col_number] if pd.notnull(df_2400[year].iloc[row_number, col_number]) else 0)
            data.append(new_row)

df_2400_long = pd.DataFrame(data, columns=columns)

display(df_2400_long)
```

```python
df_2400_long.to_csv(f'{path}Common/2400_2016-2018.csv', index=False)
```
