```python
# Загрузка библиотек
import pandas as pd
import numpy as np
```

```python
path = 'E:/IrkutskProject/Data/'

df_population = pd.read_csv(f'{path}Common/avg_population_2016-2024.csv')
df_2100 = pd.read_csv(f'{path}Common/2100_2016-2024.csv')
df_2500 = pd.read_csv(f'{path}Common/2500_2016-2024.csv')
df_2600 = pd.read_csv(f'{path}Common/2600_2016-2024.csv')
```

```python
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
columns = [
    'id',
    'name',
    'short_name',
    'year',
    'indicator',
    'numerator',
    'denominator'
]

data = []
data_2016_2018 = []

df_table_6 = pd.DataFrame(columns=columns)
df_table_6_2016_2018 = pd.DataFrame(columns=columns)

for year in years:
    new_row = []
    # 6.1 форма 33, табл.2600,   стр.1,   гр.6 /  форма 33, табл.2100,   стр.7,   гр.4 * 100%
    new_row.append('id_6.1')
    new_row.append('Охват госпитализацией впервые выявленных больных туберкулезом (%)')
    new_row.append('Охват госпит. ВВ')
    new_row.append(year)
    numerator = df_2600[(df_2600['Год'] == year) & (df_2600['Строка'] == 1) & (df_2600['Графа'] == 6)]['Значение'].iloc[0]
    denominator = df_2100[(df_2100['Год'] == year) & (df_2100['Строка'] == 7) & (df_2100['Графа'] == 4)]['Значение'].iloc[0]
    indicator = numerator*100/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
# ----------------------
    new_row = []
    # 6.2 форма 33, табл.2600,   стр.2,   гр.6 /  форма 33, табл.2500,   стр.1+4,   гр.4 * 100%
    new_row.append('id_6.2')
    new_row.append('Охват госпитализацией впервые выявленных больных туберкулезом с МБТ+ (%)')
    new_row.append('Охват госпит. ВВ+')
    new_row.append(year)
    numerator = df_2600[(df_2600['Год'] == year) & (df_2600['Строка'] == 2) & (df_2600['Графа'] == 6)]['Значение'].iloc[0]
    value_1 = df_2500[(df_2500['Год'] == year) & (df_2500['Строка'] == 1) & (df_2500['Графа'] == 3)]['Значение'].iloc[0]
    # !!!!! ошибка в данных заказчика, графа 4 заменена на 3   
    value_2 = df_2500[(df_2500['Год'] == year) & (df_2500['Строка'] == 4) & (df_2500['Графа'] == 3)]['Значение'].iloc[0]    
    denominator = value_1 + value_2
    indicator = numerator*100/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
# ----------------------
    new_row = []
    # 6.3 форма 33, табл.2600,   стр.2,   гр.3 /  форма 33, табл.2500,   стр.1+4,   гр.16 * 100%
    new_row.append('id_6.3')
    new_row.append('Охват госпитализацией контингентов больных туберкулезом с МБТ+ (%)')
    new_row.append('Охват госпит. контингент.+')
    new_row.append(year)
    numerator = df_2600[(df_2600['Год'] == year) & (df_2600['Строка'] == 2) & (df_2600['Графа'] == 3)]['Значение'].iloc[0]
    value_1 = df_2500[(df_2500['Год'] == year) & (df_2500['Строка'] == 1) & (df_2500['Графа'] == 16)]['Значение'].iloc[0]
    value_2 = df_2500[(df_2500['Год'] == year) & (df_2500['Строка'] == 4) & (df_2500['Графа'] == 16)]['Значение'].iloc[0]    
    denominator = value_1 + value_2
    indicator = numerator*100/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
# ----------------------
    new_row = []
    # 6.4 форма 33, табл.2600,   стр.1,   гр.6) / Среднегодовая численность населения Иркутской области* 100000
    new_row.append('id_6.4')
    new_row.append('Госпитализированная заболеваемость впервые выявленных больных туберкулезом (на 100 тыс. населения)')
    new_row.append('ГЗ ВВ')
    new_row.append(year)
    numerator = df_2600[(df_2600['Год'] == year) & (df_2600['Строка'] == 1) & (df_2600['Графа'] == 6)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['Всего'].iloc[0]
    indicator = numerator*100000/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
# ----------------------
    new_row = []
    # 6.5 форма 33, табл.2600,   стр.1,   гр.3) / Среднегодовая численность населения Иркутской области* 100000
    new_row.append('id_6.5')
    new_row.append('Госпитализированная заболеваемость контингентов больных туберкулезом (на 100 тыс. населения)')
    new_row.append('ГЗ контингент')
    new_row.append(year)
    numerator = df_2600[(df_2600['Год'] == year) & (df_2600['Строка'] == 1) & (df_2600['Графа'] == 3)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['Всего'].iloc[0]
    indicator = numerator*100000/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
# ----------------------
    new_row = []
    # 6.6 форма 33, табл.2600,   стр.1,   гр.7) / Среднегодовая численность населения Иркутской области детей до 14 лет  * 100000
    new_row.append('id_6.6')
    new_row.append('Госпитализированная заболеваемость впервые выявленных больных туберкулезом детей до 14 лет (на 100 тыс. населения)')
    new_row.append('ГЗ ВВ 0-14')
    new_row.append(year)
    numerator = df_2600[(df_2600['Год'] == year) & (df_2600['Строка'] == 1) & (df_2600['Графа'] == 7)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['0-14'].iloc[0]
    indicator = numerator*100000/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)
# ----------------------
    new_row = []
    # 6.7 форма 33, табл.2600,   стр.1,   гр.8) / Среднегодовая численность населения Иркутской области  подростков 15-17 лет  * 100000
    new_row.append('id_6.7')
    new_row.append('Госпитализированная заболеваемость впервые выявленных больных туберкулезом подростков 15-17 лет (на 100 тыс. населения)')
    new_row.append('ГЗ ВВ 15-17')
    new_row.append(year)
    numerator = df_2600[(df_2600['Год'] == year) & (df_2600['Строка'] == 1) & (df_2600['Графа'] == 8)]['Значение'].iloc[0]
    denominator = df_population[(df_population['Год'] == year) & (df_population['Пол'] == 'всего')]['15-17'].iloc[0]
    indicator = numerator*100000/denominator
    new_row.append(round(indicator, 2))
    new_row.append(numerator)
    new_row.append(denominator)
    data.append(new_row)
    if year <= 2018:
        data_2016_2018.append(new_row)

df_table_6 = pd.DataFrame(data, columns=columns)
df_table_6_2016_2018 = pd.DataFrame(data_2016_2018, columns=columns)
df_table_6
```
